```python
labels=weather_data[' _tempm']
```

```python
labels=labels.values
```

```python
labels=labels.reshape(-1,1)
```

```python
labels.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range = (0, 1))
weather_training_scaled = scaler.fit_transform(labels)
```

```python
features_set = []  
label_set = []  
for i in range(50, 98900):  
    features_set.append(weather_training_scaled[i-50:i,0])
    label_set.append(weather_training_scaled[i, 0])
```

```python


len(features_set)

```

```python


len(label_set)

```

```python
features_set
```

```python
features_set, label_set = np.array(features_set), np.array(label_set)
```

```python
features_set = np.reshape(features_set, (features_set.shape[0], features_set.shape[1], 1))  
```

```python
from keras.models import Sequential  
from keras.layers import Dense  
from keras.layers import LSTM  
from keras.layers import Dropout
```

```python
model = Sequential()
```

```python
model.add(LSTM(units=50, return_sequences=True, input_shape=(features_set.shape[1], 1))) 
```

```python
model.add(Dropout(0.2))  
```

```python
model.summary()
```

```python
model.add(LSTM(units=50, return_sequences=True))  
model.add(Dropout(0.2))

model.add(LSTM(units=50, return_sequences=True))  
model.add(Dropout(0.2))

model.add(LSTM(units=50))  
model.add(Dropout(0.2))  
```

```python
model.add(Dense(units = 1))  
```

```python
# model.compile(
#     optimizer=tf.keras.optimizers.RMSprop(),  # Optimizer
#     # Loss function to minimize
#     loss=keras.losses.SparseCategoricalCrossentropy(),
#     # List of metrics to monitor
#     metrics=[keras.metrics.SparseCategoricalAccuracy()],
# )
model.compile(optimizer = 'adam', loss = 'mean_squared_error',metrics =["Accuracy"])  

```

```python
model.fit(features_set, label_set, epochs = 1, validation_split = 0.2)  
```
